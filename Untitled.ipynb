```python
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
for file in os.listdir("/notebooks/br2a"):
    if file in os.listdir("/notebooks/br2b"):
        print("Found image:" + file)
        img = Image.open(os.path.join("/notebooks/br2a", file))
        img2 = Image.open(os.path.join("/notebooks/br2b", file))
        shape = (img.height, img.width)
        mask = np.zeros((shape[0], shape[1]))
        mask = np.maximum(img, img2)
        plt.imshow(mask)
        plt.show()
        cv2.imwrite(os.path.join("/notebooks/br2", file), mask)
        img.close()
        img2.close()
```
